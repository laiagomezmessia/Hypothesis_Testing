# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
import scipy.stats as stats
from scipy.stats import ttest_1samp

# Set the null hypothesis (H0) value
H0_hypothesis_value = 30  # $30/hr

# Extract the non-NaN hourly rates
hourly_rates = salaries_df['Hourly Rate'][salaries_df['Hourly Rate'].notna()]

# Perform a one-sample t-test
confidence_level = 0.95
sample_mean = hourly_rates.mean()
sample_standard_error = stats.sem(hourly_rates)
degrees_freedom = len(hourly_rates) - 1

confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

stat, pval = ttest_1samp(hourly_rates, H0_hypothesis_value)

print('confidence interval is', confidence_interval)
print('t-statistic is', stat)
print('p-value for the two-tailed test is', pval)

# interpreting the results, we can assume by the p value being much < than 0.05 that "H0 must go!". We cannot assume that 
# the hourly wage of all hourly workers is significantly different from $30/hr, but is also not equal to 30$ like H0.
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our Null hypothesis is less than or equal to, then to reject, the test statistic must be positive (rejects the <) and the p-value below 0.05 (rejects the =).

```python
# Your code here:

import scipy.stats as stats

# Set the null hypothesis (H0) value
H0_hypothesis_value = 86000  # $86,000/year

# Extract the non-NaN salaries for police force employees
police_salaries = salaries_df['Annual Salary'][(salaries_df['Department'] == 'POLICE') & salaries_df['Annual Salary'].notna()]

# Perform a one-sample t-test
confidence_level = 0.95
sample_mean = police_salaries.mean()
sample_standard_error = stats.sem(police_salaries)
degrees_freedom = len(police_salaries) - 1

confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

stat, pval = stats.ttest_1samp(police_salaries, H0_hypothesis_value)

print('confidence interval is', confidence_interval)
print('t-statistic is', stat)
print('p-value for the one-tailed test is', pval / 2)  # Divide by 2 for one-tailed test

# Check if the test is significant at the 0.05 level
if pval / 2 < 0.05:
    print("Reject the null hypothesis. Salaries in the police force are higher than $86,000/year.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that salaries are higher than $86,000/year.")
```

The workers from the department with the most hourly workers have complained that their hourly wage is less than $35/hour. Using a one sample t-test, test this one-sided hypothesis at the 95% confidence level.

```python
# Set the null hypothesis (H0) value
H0_Hourly_rate = 35  # $35/hr

# Create a subset of data for hourly workers from the department with the most hourly workers
data_hourly = salaries_df[(salaries_df['Salary or Hourly'] == 'Hourly') & (salaries_df['Department'] == department_with_most_hourly_workers)]

# Perform a one-sample t-test for a one-tailed (left-tailed) test
confidence_level = 0.95
sample_mean = data_hourly['Hourly Rate'].mean()
sample_standard_error = stats.sem(data_hourly['Hourly Rate'])
degrees_freedom = len(data_hourly) - 1

confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

# Calculate the t-statistic and p-value for the one-tailed test
stat, pval = ttest_1samp(data_hourly['Hourly Rate'], H0_Hourly_rate)

# Check if we can reject the null hypothesis
if stat < 0 and pval / 2 < 0.05:  # Divide p-value by 2 for one-tailed test
    print("We reject the null hypothesis: Hourly wage is significantly less than $35/hour.")
else:
    print("We fail to reject the null hypothesis: Hourly wage is not significantly less than $35/hour.")

print('Confidence interval is', confidence_interval)
print('t-statistic is', stat)
print('p-value for the one-tailed test is', pval / 2)  # Divide p-value by 2 for one-tailed test
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here:
# we can find a confidence interval for the population based on our sample
import scipy.stats

hourly_wage_data = data_new['Hourly Rate']


confidence_level = 0.95
#confidence_level = 1  # WHAT HAPPENS NOW?
degrees_freedom = len(hourly_wage_data) - 1  # or, for large values of sample size, just use that
sample_mean = np.mean(hourly_wage_data)
# note that we use the standard error of the sample 
# as an estimate of the standard error of the population (which is used in the theoretical formula)
sample_standard_error = scipy.stats.sem(hourly_wage_data) # sem = standard error of the mean = std(patients)/sqrt(degrees_freedom)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)


print( 'sample mean is ', sample_mean, '.' )
print( 'confidence interval is ', confidence_interval, '.' )
```
